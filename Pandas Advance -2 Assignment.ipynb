## Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
### Answer:-
### both loc and iloc are used to access data from a pandas DataFrame in Python, but they differ in how they select the data.
### loc is label-based and allows you to access rows and columns by labels, whereas iloc is integer-based and allows you to access rows and columns by integer index.
### loc is used to access rows and columns by labels. It takes two arguments: the row label(s) and column label(s). The row and column labels can be a single label or a list of labels.

### iloc is used to access rows and columns by integer index. It takes two arguments: the row index(es) and column index(es). The row and column indexes can be a single integer or a list of integers.

### loc includes the last element when using a range, while iloc does not. For example, df.loc[1:3] will return rows with labels 1, 2, and 3 (inclusive), while df.iloc[1:3] will return rows with indexes 1 and 2 (exclusive).

### loc can accept boolean arrays to filter data by condition, while iloc can only accept integer arrays.

### loc is slower than iloc for accessing data because it involves label-based lookups.

## Q6. What do you understand about the windows function in pandas and list the types of windows functions?
## Answers:- 

### a window function is a way to perform calculations over a specified window of rows in a DataFrame or a Series. The window can be defined using a sliding or expanding window, and the calculation is performed on the data within the window.

### There are several types of window functions in pandas, including:

### 1. Rolling window functions: These functions calculate values over a rolling window of fixed size. The window "rolls" over the DataFrame or Series and the calculation is performed on the data within the window. Some examples of rolling window functions are rolling(), rolling_mean(), rolling_sum(), rolling_min(), and rolling_max().

### 2. Expanding window functions: These functions calculate values over an expanding window that includes all rows up to the current row. The window size increases with each row, and the calculation is performed on the data within the window. Some examples of expanding window functions are expanding(), expanding_mean(), expanding_sum(), expanding_min(), and expanding_max().

### 3. Exponentially weighted window functions: These functions calculate exponentially weighted statistics over a rolling window of fixed size. The weights of each value in the window decrease exponentially as they move further away from the current row. Some examples of exponentially weighted window functions are ewm(), ewm_mean(), ewm_min(), and ewm_max().

### 4. Other window functions: There are several other window functions available in pandas, such as shift(), diff(), and pct_change(). These functions do not calculate values over a fixed window size but are still considered window functions because they operate on a set of rows defined relative to the current row.

## Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

```python

# Prompt the user to enter the two dates

date1 = input("Enter the first date in YYYY-MM-DD format: ")
date2 = input("Enter the second date in YYYY-MM-DD format: ")

# Convert the input strings to datetime objects

date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# Calculate the difference between the two dates

delta = date2 - date1

# Display the result in days, hours, and minutes

print("The difference between the two dates is:")
print("{} days".format(delta.days))
print("{} hours".format(delta.seconds // 3600))
print("{} minutes".format((delta.seconds // 60) % 60))
```

## Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

```python
filepath = input('Enter the filepath: ')
column = input('Enter the column name: ')
order = input('Enter category order: ')

df = pd.read_csv(filepath)

df[column].astype('category')

if order=='ascending':
    df=df.sort_values(by = column, ascending= True)
elif order=='descending':
    df=df.sort_values(by = column, ascending= False)

df
```

## Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.

```python
pd.read_csv("product_sales.csv")
```

```python

filepath = input('Enter filepath: ')

df = pd.read_csv(filepath)
df['Date'] = pd.to_datetime(df['Date'])
grouped = df.groupby(['Product Category', 'Date']).sum()
pivot = grouped.unstack(level=0)
pivot.plot(kind='bar', stacked=True)

```

## Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and displays the results in a table.

### The program should do the following:

### Prompt the user to enter the file path of the CSV file containing the student data
### Read the CSV file into a Pandas DataFrame
### Calculate the mean, median, and mode of the test scores using Pandas toolsR
### Display the mean, median, and mode in a table.

```python
filepath = input('Enter Filepath: ')

df = pd.read_csv(filepath)

mean = df['Test Score'].mean()
median = df['Test Score'].median()
mode = df['Test Score'].mode()

print("+----------+----------+")
print("Statistic | Value")
print(f"Mean:    {mean}")
print(f"Median:    {median}")
print(f"Mode:    {mode}")
print("+----------+----------+")
```
